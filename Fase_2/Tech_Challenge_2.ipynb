```python
import pandas as pd
```

```python
df = pd.read_csv('/content/Dados Históricos - Ibovespa 02-01-2020_a_01-07-2025.csv', delimiter=',')
```

```python
df.dtypes
```

```python
df['Data'] = pd.to_datetime(df['Data'], format='%d.%m.%Y')
df = df.sort_values('Data')
df.head()
```

```python
df.shape
```

```python
df.dropna()
```

```python
df.shape
```

```python
df['Variacao 1/0'] = (df['Último'].shift(-1) > df['Último']).astype(int)
df.head(30)
```

```python
df['mm_5'] = df['Último'].rolling(window=5).mean()
df['mm_10'] = df['Último'].rolling(window=10).mean()
df.head(20)
```

```python
df['retorno_diario'] = df['Último'].pct_change()
df.head(20)
```

```python
df['retorno_acumulado'] = (df['Último'] / df['Último'].iloc[0]) - 1
df.head()
```

```python
df['volatilidade_5d'] = df['retorno_diario'].rolling(window=5).std()
df['vol_21d'] = df['retorno_diario'].rolling(window=21).std()
df['vol_historica_anual'] = df['vol_21d'] * (252**0.5)
df['vol_historica'] = df['retorno_diario'].std()
df.head(22)
```

```python
df = df.fillna(0)
df.head()
```

```python
df.to_csv('Ibovespa 02-01-2020_a_01-07-2025_ajustado.csv', index=False)
```
